```python
import torch
import torch.nn as nn

class ResBlock(nn.Module):
    """Basic Residual Block"""
    def __init__(self, channels):
        super(ResBlock, self).__init__()
        self.conv1 = nn.Conv2d(channels, channels, 3, stride=1, padding=1)  # First convolution
        self.bn1 = nn.BatchNorm2d(channels)  # Batch normalization
        self.relu = nn.ReLU(inplace=True)  # ReLU activation
        self.conv2 = nn.Conv2d(channels, channels, 3, stride=1, padding=1)  # Second convolution
        self.bn2 = nn.BatchNorm2d(channels)  # Batch normalization

    def forward(self, x):
        residual = x  # The skip connection (input to the block)
        out = self.conv1(x)  # Apply first convolution
        out = self.bn1(out)  # Apply batch normalization
        out = self.relu(out)  # Apply ReLU activation
        out = self.conv2(out)  # Apply second convolution
        out = self.bn2(out)  # Apply batch normalization
        out += residual  # Add the residual connection to the output
        return self.relu(out)  # Apply ReLU activation after adding residual

class Generator(nn.Module):
    def __init__(self, z_dim=128, img_channels=3):
        super(Generator, self).__init__()
        # Latent vector input (z_dim = 128 for example, size: [batch_size, 128, 1, 1])
        self.initial_conv = nn.ConvTranspose2d(z_dim, 512, 4, 1, 0)  # Upsample to (batch_size, 512, 4, 4)
        
        self.resblock1 = ResBlock(512)
        self.resblock2 = ResBlock(512)
        self.resblock3 = ResBlock(512)
        
        # Upsample to higher resolution (e.g., 32x32)
        self.upsample1 = nn.ConvTranspose2d(512, 256, 4, 2, 1)  # Upsample to (batch_size, 256, 8, 8)
        self.upsample2 = nn.ConvTranspose2d(256, 128, 4, 2, 1)  # Upsample to (batch_size, 128, 16, 16)
        self.upsample3 = nn.ConvTranspose2d(128, 64, 4, 2, 1)  # Upsample to (batch_size, 64, 32, 32)
        self.upsample4 = nn.ConvTranspose2d(64, 32, 4, 2, 1)  # Upsample to (batch_size, 32, 64, 64)
        # Final output (batch_size, img_channels, 64, 64)
        self.final_conv = nn.Conv2d(32, img_channels, 3, stride=1, padding=1)

    def forward(self, z):
        x = self.initial_conv(z)  # Shape: [batch_size, 512, 4, 4]
        #print(x.size())
        x = self.resblock1(x)  # Shape: [batch_size, 512, 4, 4]
        x = self.resblock2(x)  # Shape: [batch_size, 512, 4, 4]
        x = self.resblock3(x)  # Shape: [batch_size, 512, 4, 4]
        
        x = self.upsample1(x)  # Shape: [batch_size, 256, 8, 8]
        x = self.upsample2(x)  # Shape: [batch_size, 128, 16, 16]
        x = self.upsample3(x)  # Shape: [batch_size, 64, 32, 32]
        x = self.upsample4(x)
        x = self.final_conv(x)  # Shape: [batch_size, img_channels, 32, 32]
        #print(x.size())
        return torch.tanh(x)  # Final output image
```

```python
def get_training_device():
    # Use the GPU if possible
    if torch.cuda.is_available():
        return torch.device('cuda')
    # Otherwise use the CPU :-(
    return torch.device('cpu')

def to_device(data, device):
    # This moves the tensors to the device (GPU, CPU)
    if isinstance(data, (list,tuple)):
        return [to_device(x, device) for x in data]
    return data.to(device, non_blocking=True)

class DeviceDataLoader():
    """Wrap a dataloader to move data to a device"""
    def __init__(self, dataloader, device):
        self.dataloader = dataloader
        self.device = device
        
    def __iter__(self):
        """Yield a batch of data after moving it to device"""
        for b in self.dataloader: 
            yield to_device(b, self.device)

    def __len__(self):
        """Number of batches"""
        return len(self.dataloader)
```

```python
def compute_gradient_penalty(discriminator, real_samples, fake_samples):
    alpha = torch.rand(real_samples.size(0), 1, 1, 1, device=device)  # Random weight
    interpolates = (alpha * real_samples + (1 - alpha) * fake_samples).requires_grad_(True)
    disc_interpolates = discriminator(interpolates)

    gradients = torch.autograd.grad(
        outputs=disc_interpolates,
        inputs=interpolates,
        grad_outputs=torch.ones_like(disc_interpolates),
        create_graph=True,
        retain_graph=True,
        only_inputs=True
    )[0]

    gradients = gradients.view(gradients.size(0), -1)
    gradient_penalty = ((gradients.norm(2, dim=1) - 1) ** 2).mean()
    return gradient_penalty
```

```python
def train_discriminator(real_pokemon, disc_optimizer):
    # Reset the gradients for the optimizer
    disc_optimizer.zero_grad()

    # Création d'une seed pour le générateur
    latent_batch = torch.randn(batch_size, seed_size, 1, 1, device=device)

    # Generation de faux pokemons
    fake_pokemon = generator(latent_batch)

    # Prédictions du discriminateur
    real_predictions = discriminator(real_pokemon)
    gen_predictions = discriminator(fake_pokemon)
    #Choix de la loss
    choice='cross_entropy'
    #choice='wgan'
    # real_targets = torch.zeros(real_pokemon.size(0), 1, device=device) # All of these are real, so the target is 0.
    if choice=='cross_entropy':
        real_targets = torch.rand(real_pokemon.size(0), 1, device=device) * (0.1 - 0) + 0 # Add some noisy labels to make the discriminator think harder.
        real_loss = F.binary_cross_entropy(real_predictions, real_targets) # Can do binary loss function because it is a binary classifier
        # Train on the generator's current efforts to trick the discriminator
        
        # gen_targets = torch.ones(fake_pokemon.size(0), 1, device=device)
        gen_targets = torch.rand(fake_pokemon.size(0), 1, device=device) * (1 - 0.9) + 0.9 # Add some noisy labels to make the discriminator think harder.
        gen_loss = F.binary_cross_entropy(gen_predictions, gen_targets)
        total_loss = real_loss + gen_loss
        total_loss.backward()
        torch.nn.utils.clip_grad_norm_(discriminator.parameters(), max_norm=0.01)
        disc_optimizer.step()
    elif choice=='wgan':
        
        loss = (real_predictions.mean()-gen_predictions.mean())

        print("real     =",real_predictions.mean())
        print("generated=",gen_predictions.mean())
        #print(gen_predictions)
        real_score = torch.mean(real_predictions).item() # How well does the discriminator classify the real pokemon? (Higher score is better for the discriminator)
        gen_score = torch.mean(gen_predictions).item() # How well did the discriminator classify the fake pokemon? (Lower score is better for the discriminator)
        # Update the discriminator weights
        gp = compute_gradient_penalty(discriminator, real_pokemon, fake_pokemon)
        loss = -real_predictions.mean()+gen_predictions.mean() + 10 * gp  # λ=10
        total_loss = loss
        total_loss.backward()
        #torch.nn.utils.clip_grad_norm_(discriminator.parameters(), max_norm=0.01)
        disc_optimizer.step()
    return total_loss.item(), real_score, gen_score
```

```python
def train_generator(gen_optimizer):
    # Clear the generator gradients
    gen_optimizer.zero_grad()
    
    # Generate some fake pokemon
    latent_batch = torch.randn(batch_size, seed_size, 1, 1, device=device)

    fake_pokemon = generator(latent_batch)
    
    # Test against the discriminator
    disc_predictions = discriminator(fake_pokemon)

    #Choix de la loss
    choice='cross_entropy'
    #choice='wgan'

    if choice=='cross_entropy':
        targets = torch.zeros(fake_pokemon.size(0), 1, device=device) # We want the discriminator to think these images are real.
        loss = F.binary_cross_entropy(disc_predictions, targets) # How well did the generator do? (How much did the discriminator believe the generator?)
    elif choice=='wgan':
        loss = -disc_predictions.mean()
        #print(loss.item())
    # Update the generator based on how well it fooled the discriminator
    loss.backward()
    
    gen_optimizer.step()
    
    # Return generator loss
    return loss.item()
```

```python
def train_discriminator(real_pokemon, disc_optimizer):
    disc_optimizer.zero_grad()

    latent_batch = torch.randn(real_pokemon.size(0), seed_size, 1, 1, device=device)
    fake_pokemon = generator(latent_batch).detach()  # Detach to avoid updating generator

    real_predictions = discriminator(real_pokemon)
    gen_predictions = discriminator(fake_pokemon)

    choice = 'wgan'
    choice = 'cross_entropy'
    use_gradient_penalty=True
    if choice == 'cross_entropy':
        real_targets = torch.rand(real_pokemon.size(0), 1, device=device) * 0.1  # Noisy labels
        real_loss = F.binary_cross_entropy(real_predictions, real_targets)

        gen_targets = torch.rand(fake_pokemon.size(0), 1, device=device) * 0.1 + 0.9
        gen_loss = F.binary_cross_entropy(gen_predictions, gen_targets)

        total_loss = real_loss + gen_loss
    elif choice == 'wgan':

        gp = compute_gradient_penalty(discriminator, real_pokemon, fake_pokemon)
        print("real=",real_predictions.mean() )
        print("fake=",gen_predictions.mean())
        print("gp=",gp)
        loss = -real_predictions.mean() + gen_predictions.mean() + 500 * gp  # λ=10
        total_loss = loss

    total_loss.backward()
    disc_optimizer.step()

    # Clip weights (only if NOT using gradient penalty)
    if choice == 'wgan' and not use_gradient_penalty:
        for p in discriminator.parameters():
            p.data.clamp_(-0.01, 0.01)

    return total_loss.item(), real_predictions.mean().item(), gen_predictions.mean().item()


def train_generator(gen_optimizer):
    gen_optimizer.zero_grad()

    latent_batch = torch.randn(batch_size, seed_size, 1, 1, device=device)
    fake_pokemon = generator(latent_batch)

    disc_predictions = discriminator(fake_pokemon)

    choice = 'wgan'
    choice = 'cross_entropy'
    if choice == 'cross_entropy':
        targets = torch.zeros(fake_pokemon.size(0), 1, device=device)  # Want discriminator to think it's real
        loss = F.binary_cross_entropy(disc_predictions, targets)
    elif choice == 'wgan':
        loss = -disc_predictions.mean()  # Flip sign

    loss.backward()
    gen_optimizer.step()

    return loss.item()
```

```python
def save_results(index, latent_batch, show=True):
    # Generate fake pokemon
    fake_pokemon = generator(latent_batch)
    print(fake_pokemon.size())
    # Make the filename for the output
    fake_file = "result-image-{0:0=4d}.png".format(index)
    
    # Save the image
    save_image(denorm(fake_pokemon), os.path.join(RESULTS_DIR, fake_file), nrow=8)
    print("Result Saved!")
    show=True
    if show:
        fig, ax = plt.subplots(figsize=(8, 8))
        ax.set_xticks([]); ax.set_yticks([])
        ax.imshow(make_grid(fake_pokemon.cpu().detach(), nrow=8).permute(1, 2, 0))
```

```python
from tqdm.notebook import tqdm
import torch.nn.functional as F

# Static generation seed batch
fixed_latent_batch = torch.randn(64, seed_size, 1, 1, device=device)

def train(epochs, learning_rate, start_idx=1):
    # Empty the GPU cache to save some memory
    torch.cuda.empty_cache()
    
    # Track losses and scores
    disc_losses = []
    disc_scores = []
    gen_losses = []
    gen_scores = []
    
    # Create the optimizers
    disc_optimizer = torch.optim.Adam(discriminator.parameters(), lr=learning_rate, betas=(0.5, 0.9))
    gen_optimizer = torch.optim.Adam(generator.parameters(), lr=learning_rate, betas=(0.5, 0.9))
    
    # Run the loop
    for epoch in range(epochs):
        # Go through each image
        for real_img, _ in tqdm(dev_dataloader):
            # Train the discriminator
            disc_loss, real_score, gen_score = train_discriminator(real_img, disc_optimizer)

            # Train the generator
            gen_loss = train_generator(gen_optimizer)
        
        # Collect results
        disc_losses.append(disc_loss)
        disc_scores.append(real_score)
        gen_losses.append(gen_loss)
        gen_scores.append(gen_score)
        
        # Print the losses and scores
        print("Epoch [{}/{}], gen_loss: {:.4f}, disc_loss: {:.4f}, real_score: {:.4f}, gen_score: {:.4f}".format(
            epoch+start_idx, epochs, gen_loss, disc_loss, real_score, gen_score))
        
        # Save the images and show the progress
        save_results(epoch + start_idx, fixed_latent_batch, show=False)
    
    # Return stats
    return disc_losses, disc_scores, gen_losses, gen_scores
```

```python
def initialize_weights(model):
    for layer in model:
        if isinstance(layer, nn.Conv2d):
            init.xavier_uniform_(layer.weight)  # Apply Xavier Uniform initialization
            if layer.bias is not None:
                init.zeros_(layer.bias)  # Initialize biases to zeros
        elif isinstance(layer, nn.Linear):
            init.xavier_uniform_(layer.weight)
            if layer.bias is not None:
                init.zeros_(layer.bias)

```
